```python
import seaborn as sns
import pandas as pd
import numpy as np

df = sns.load_dataset('titanic')
df.head(10)
```

```python
df[(df['sex'] == 'female') & (df['age'] < 18) & (df['survived'] == 1)]
```

```python
median_age = df['age'].median()
df['age'] = df['age'].fillna(median_age)
```

```python
def age_group(age):
    if age < 18:
        return 'Child'
    elif age <= 60:
        return 'Adult'
    else:
        return 'Senior'

df['age_group'] = df['age'].apply(age_group)

df[['age', 'age_group']].head()
```
