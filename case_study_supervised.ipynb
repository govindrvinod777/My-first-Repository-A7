```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/Training_data.csv")
```

```python
data
```

```python
data.isna().sum()
```

```python
for i in data.select_dtypes(include=['number']):
    plt.figure(figsize=(10,5))
    plt.boxplot(data[i].dropna())
    plt.title(i)
```

```python
num_cols = pd.concat([data.select_dtypes(include='float64'), data.select_dtypes(include='int64')], axis=1)
num_cols
```

```python
num_cols.hist(figsize=(10,10))
```

```python
data.fillna({'MonthlyCharges': data['MonthlyCharges'].median()}, inplace=True)
```

```python
data.fillna({'gender': data['gender'].mode()[0]}, inplace=True)
```

```python
data.fillna({'MonthlyCharges': data['MonthlyCharges'].mean()}, inplace=True)
```
