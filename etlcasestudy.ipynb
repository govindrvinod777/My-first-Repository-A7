```python
import pandas as pd
import numpy as np
```

```python
data =pd.read_csv("/content/all_month.csv")
#Load the dataset into a DataFrame
data = pd.DataFrame(data)
```

```python

#Identify number of records and unique locations
number_of_records =len(data)
number_of_records
```

```python
unique_location = data['place'].nunique()
unique_location
```

```python

data.head(5)
```

```python
data.columns
```

```python

data['time']=pd.to_datetime(data['time'])
data.info()
```

```python

print(data.isnull().sum())
```

```python
data = data.dropna(subset=['latitude','longitude','mag'])
```

```python
data[data['mag'] >= 4.0]
```

```python

data['day_of_week'] =data['time'].dt.day_name()
data.columns
```

```python
conditions =[
    (data['mag']<4.0),
    (data['mag']>=4.0)&(data['mag']<=6.0),
    (data['mag']>6.0)
]
choices=['Low','Moderate','High']
data['severity_level']=np.select(conditions,choices,default='unknown')
```

```python
data.columns
```

```python

earthquake_count=data['place'].value_counts()
print(earthquake_count)
```

```python
daily_stats = data.groupby('day_of_week').agg(
    average_magnitude=('mag', 'mean'),
    max_depth=('depth', 'max')
)
display(daily_stats)
```

```python
data.to_csv('cleaned_earthquakes.csv')
daily_stats.to_csv('earthquake_summary.csv')
```

##A short explanation of your process

**Extraction**: I loaded the earthquake data from the provided CSV file into a pandas DataFrame.

**Transformation**: I performed several data cleaning and feature engineering steps:
Converted the 'time' column to datetime objects.
Dropped rows with missing values in critical columns ('latitude', 'longitude', 'mag').
Filtered for earthquakes with a magnitude of 4.0 or greater
Created a 'day_of_week' column from the timestamp.
Created a 'severity_level' column based on the magnitude.
Counted the number of earthquakes per place.
Computed the average magnitude and maximum depth for each day of the week.


**Loading**: I saved the cleaned dataset and the daily statistics to CSV files and also stored them in tables within a SQLite database.
